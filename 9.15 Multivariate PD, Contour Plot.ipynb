```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("Iris.csv")

iris_setosa = iris.loc[iris["Species"] == "Iris-setosa"] #taking all points corresponding to setosa
```

```python
#2D Density plot, contors-plot
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris_setosa, kind="kde");
plt.show();

#x-axis=PetalLengthCm, y-axis=PetalWidthCm
#If we see the scatter plot for PL v/s PW, we see there are more no. of points in the centre and as we move away from the centre, the no. of points are decreasing, refer image 1
#In 1D density plot/pdf plot, if there were more no. of points the height would be more

#Imagine a hill, if the no. of points are higher, the hill's peak will be higher, else lower , refer image 2
#Since there are more no of points where x-axis=1.4 and y-axis=0.2, we can think of it as a hill coming towards us, which has a high height
#the circles are called Contour and the plot is called Contour Density Plot where x-axis=PetalLengthCm, y-axis=PetalWidthCm and height is the prob of finding those points
#all the points inside a contour/circle have the same height on the prob hill

#the 2 distributions on the x and y axis represent the pdfs of PL and PW and the combo of both is a 2D plot

#This plot is also called Multi Variate Density Plot
```
