```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("Iris.csv")
```

```python
# A violin plot has the bemefits of PDF, Histogram, Box Plots
#and simplifies them

# Denser regions of the data are fatter, and sparser ones thinner 
#in a violin plot
sns.violinplot(x='Species',y='PetalLengthCm', data=iris, size=8)
plt.grid()
plt.show()
```

```python
#The thick black area is a box plot (25, 50, 75 percentiles and whiskers)
#The curves are the PDFs plotted on the side
#looks like a Bell Curve
#preferred plots: box, violin
```
