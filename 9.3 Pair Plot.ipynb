```python
Pair Plots can be used to visualize/compare multi dimension(ie. data with multiple features) data
We have 4 variables here 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'. We saw a plot of ('SepalLengthCm', 'SepalWidthCm') in 9.1
How many such pair exists? 4C2 = 6 pairs or plots exists, ie one 2D plot of 2 variables and there are a total of 4 variables
Pairs:
(SL, SW)    (SW, PL)
(SL, PL)    (SW, PW)
(SL, PW)    (PL, PW)
We will visualize these 6 2D plots, instead of a single 4D plot to understand the data
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("Iris.csv")
```

```python
#pairwise scatter plot: Pair Plot
#Disadvantages:
1. Canmot be used only when no. of features are high
2. Cannot visualize higher dimensional patterns in 3D, 4D
3. Only possible to view 2D patterns
```

```python
df = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']]
df.columns
```

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(df, hue="Species",size=3)
plt.show()
```
