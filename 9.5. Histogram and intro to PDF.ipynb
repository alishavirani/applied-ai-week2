```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("Iris.csv")
```

```python
#what about 1d Scatter plot using just 1 feature?
#1D scatter plot of petal length

iris_setosa = iris.loc[iris["Species"] == "Iris-setosa"] #taking all points corresponding to setosa
iris_virginica = iris.loc[iris["Species"] == "Iris-virginica"] #taking all points corresponding to virginica
iris_versicolor = iris.loc[iris["Species"] == "Iris-versicolor"] #taking all points corresponding to versicolor

# print(iris_setosa)
# print(np.zeros_like(iris_setosa["PetalLengthCm"]))

plt.plot(iris_setosa["PetalLengthCm"], np.zeros_like(iris_setosa["PetalLengthCm"])) #x-axis petal lengths of iris_setosa, y-axis of all zeros corresponding to iris_setosa
plt.plot(iris_virginica["PetalLengthCm"], np.zeros_like(iris_virginica["PetalLengthCm"]))
plt.plot(iris_versicolor["PetalLengthCm"], np.zeros_like(iris_versicolor["PetalLengthCm"]))

plt.xlabel("PetalLengthCm")
plt.show()

#Disadvantages of 1D scatter plot:
#Very hard to make sense as points are overlapping a lot
#cannot count no. of points in a particular region
```
