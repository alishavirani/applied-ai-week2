```python
#Unvariate analysis = One variable analysis
Which of the 4 variables out of SL, SW, PW, PL help distinguish the 3 flowers better
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

iris = pd.read_csv("Iris.csv")
```

```python
df = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']]
```

```python
#understanding which of the 4 features help classify better
```

```python
sns.FacetGrid(df, hue="Species", size=5).map(sns.displot, "PetalLengthCm").add_legend();

plt.show()
```

```python
#Above has well seperation between setosa and rest, ie setosa is well seperated
#there is some seperation between versicolor and virginica, because there is a small overlap
#best case would be all 3 well seperated
```

```python
sns.FacetGrid(df, hue="Species", size=5).map(sns.displot, "PetalWidthCm").add_legend();

plt.show()
```

```python
#setosa and versicolor are not much seperated
#there is some seperation between versicolor and virginica, because there is a small overlap
#petallength does a better job here, bcoz setosa is very well seperated
#farther the distribution, the better
```

```python
sns.FacetGrid(df, hue="Species", size=5).map(sns.displot, "SepalLengthCm").add_legend();

plt.show()
```

```python
#Massive overlap of all 3
#PL > PW > SL (order of betterment)
```

```python
sns.FacetGrid(df, hue="Species", size=5).map(sns.displot, "SepalWidthCm").add_legend();

plt.show()
```

```python
#Very bad overlapping, almost full overlaps on all 3

PL > PW >> SL >> SW #scale of betterment

#If choosing 1 feature, we would choose PL, if 2 then PW (can see pair plots to identify which feature is better)
```
